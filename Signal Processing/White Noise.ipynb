 # White Noise (Ruido Blanco)

```python
import numpy as np
import matplotlib.pyplot as plt
#from matplotlib import cm
%matplotlib inline
plt.rcParams['font.size'] = '12'
```

```python
# Ruido Blanco
N = 1000

x1 = 2*np.random.randint(2,size=N) - 1
x2 = np.random.randn(N)
x3 = (2*np.random.random(N) - 1)*np.sqrt(3)

plt.figure(figsize=[12,12])
plt.subplot(3,1,1)
plt.stem(x1,basefmt="")
plt.grid('on')
#plt.xlabel('$n$')
plt.xlim([0,100])
plt.ylim([-2.5,2.5])
plt.title('Ruido Blanco Bernoulli')

plt.subplot(3,1,2)
plt.stem(x2,basefmt="")
plt.grid('on')
#plt.xlabel('$n$')
plt.xlim([0,100])
plt.ylim([-2.5,2.5])
plt.title('Ruido Blanco Gaussiano')

plt.subplot(3,1,3)
plt.stem(x3,basefmt="")
plt.grid('on')
plt.xlabel('$n$')
plt.xlim([0,100])
plt.ylim([-2.5,2.5])
plt.title('Ruido Blanco Distribuci√≥n Uniforme')

```

```python
# Calculando las distribuciones correspondientes

xrange = [-3,3]

plt.figure(figsize=[4,12])

Nb = 13
dx = (xrange[1] - xrange[0]) / (Nb-1)
bins = xrange[0]-dx/2 + np.arange(Nb+1)*dx

plt.subplot(3,1,1)
plt.hist(x1,bins=bins,rwidth=0.8,align='mid',ec='k', weights=np.ones(N)/N)
plt.grid('on')

plt.subplot(3,1,2)
plt.hist(x2,bins=bins,rwidth=0.8,align='mid',ec='k',density=True)
plt.grid('on')

plt.subplot(3,1,3)
plt.hist(x3,bins=bins,rwidth=0.8,align='mid',ec='k',density=True)
plt.grid('on')
```

```python
#  Calculando las varianzas
var1 = np.var(x1)
var2 = np.var(x2)
var3 = np.var(x3)
print('Varianza 1: %0.4f'%(var1))
print('Varianza 2: %0.4f'%(var2))
print('Varianza 3: %0.4f'%(var3))
```

```python
# Calculando la correlaci√≥n
corr1 = np.correlate(x1,x1,mode='full')
corr2 = np.correlate(x2,x2,mode='full')
corr3 = np.correlate(x3,x3,mode='full')

plt.figure(figsize=[8,6])
plt.plot(corr1/N)
plt.plot(corr2/N)
plt.plot(corr3/N)
plt.grid('on')
plt.legend(['Bernoulli','Gaussiano','Uniforme'])
```
