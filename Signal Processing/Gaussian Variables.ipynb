# Variables Gaussianas

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

plt.rcParams['font.size'] = '12'
%matplotlib inline
```

```python
# Función de Densidad Gaussiana Simple
N = 101
x = np.linspace(-3.5,3.5,N)
fx = np.exp(-x**2/2)/np.sqrt(2*np.pi)

plt.figure()
plt.plot(x,fx)
plt.grid('on')
plt.xlabel('$x$')
plt.ylabel('$f_X(x)$')
plt.xlim([-3.5,3.5])
```

```python
# Función de Densidad Gaussiana Conjunta
N = 101
x = np.linspace(-3.5,3.5,N)
y = x
xx, yy = np.meshgrid(x,y)
fxy = np.exp(-(xx**2+yy**2)/2)/(2*np.pi)

fig = plt.figure(figsize=[8,6])
ax = plt.axes(projection='3d')
surf = ax.plot_surface(xx,yy,fxy,cmap=cm.coolwarm)
plt.grid('on')
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_zlabel('$f_{XY}(x,y)$')
ax.set_xlim([-3.5,3.5])
ax.set_ylim([-3.5,3.5])
ax.set_zlim([0.,np.max(fxy)*1.1])

fig.colorbar(surf,location='left',shrink=0.8);
```
