# Señales Determinísticas y Aleatorias

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.size'] = '12'
#%matplotlib inline
```

```python
# Comparando señales determinísticas y aleatorias
N = 120
fn = 0.05

n = np.arange(N)
x1 = 2*np.sin(2*np.pi*fn*n)
x2 = 2*np.sin(2*np.pi*fn*n) + np.random.randn(N)

plt.figure(figsize=[12,4])
plt.plot(x1,'o-')
plt.grid('on')
plt.xlabel('$n$')
plt.ylabel('$x_1[n]$')

plt.figure(figsize=[12,4])
plt.plot(x2,'ro-')
plt.grid('on')
plt.xlabel('$n$')
plt.ylabel('$x_2[n]$')
```

```python
# Comparando señales determinísticas y aleatorias en el dominio de la frecuencia

Xw1 = np.fft.fft(x1)
Xw1 = np.fft.fftshift(Xw1)

Xw2 = np.fft.fft(x2)
Xw2 = np.fft.fftshift(Xw2)

w = 2*np.pi*(np.arange(N) - N/2)/N

plt.figure(figsize=[12,4])
plt.stem(w,np.abs(Xw1), basefmt=' ')
plt.grid('on')
plt.xlabel('$\omega$ [rad]')
plt.ylabel('$X_1(\omega)$')

plt.figure(figsize=[12,4])
plt.stem(w,np.abs(Xw2),'r', markerfmt='ro', basefmt=' ')
plt.grid('on')
plt.xlabel('$\omega$ [rad]')
plt.ylabel('$X_2(\omega)$')
```
