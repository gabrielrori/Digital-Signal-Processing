# Transformada Z - Secuencia de longitud finita

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
#%matplotlib widget
%matplotlib inline
```

```python
# Ejemplo: X(z) = x[0] + x[1] z^{-1}

x = np.array([1,1])

rz = np.linspace(-3,3,100)
iz = np.linspace(-3,3,100)

rr, ii = np.meshgrid(rz,iz)
zz = rr+1j*ii

Xz = x[0] + x[1]*zz**(-1)
```

```python
fig = plt.figure(figsize=[12,4])
(ax1,ax2) = fig.subplots(1,2)
im = ax1.pcolormesh(rz,iz,np.abs(Xz),cmap=cm.jet,shading='auto')
ax1.set_xlabel('$\mathrm{Re}\{z\}$')
ax1.set_ylabel('$\mathrm{Im}\{z\}$')
ax1.set_title('$|X(z)|$')
ax1.set_aspect(1)
ax1.grid('on',linestyle='--')
im.set_clim(0,10)
fig.colorbar(im,ax=ax1)

circ = plt.Circle((0, 0), radius=1, edgecolor='k', facecolor='None')
ax1.add_artist(circ)

im = ax2.pcolormesh(rz,iz,np.angle(Xz),cmap=cm.coolwarm,shading='auto')
ax2.set_xlabel('$\mathrm{Re}\{z\}$')
ax2.set_ylabel('$\mathrm{Im}\{z\}$')
ax2.set_title('$\\angle X(z)$')
ax2.set_aspect(1)
ax2.grid('on',linestyle='--')
im.set_clim(-np.pi,np.pi)
fig.colorbar(im,ax=ax2)

circ = plt.Circle((0, 0), radius=1, edgecolor='k', facecolor='None')
ax2.add_artist(circ)
```

```python
# Calculando DTFT{x[n]}
omg = 2*np.pi*(np.linspace(0,1,101)-0.5)
z = np.exp(1j*omg)
Xomg = x[0] + x[1]*z**(-1)
```

```python
fig = plt.figure(figsize=[12,4])

plt.subplot(1,2,1)
plt.plot(omg,np.abs(Xomg))
plt.xlabel('$\omega$ [rad]')
plt.title('$|X(e^{j\omega})|$')
plt.grid()

plt.subplot(1,2,2)
plt.plot(omg,np.angle(Xomg))
plt.xlabel('$\omega$ [rad]')
plt.title('$\\angle X(e^{j\omega})$')
plt.grid()
```

```python
# Calculando la FFT
N = 16
Xfft = np.fft.fft(x,N)
Xfft = np.fft.fftshift(Xfft)
wn = 2*np.pi*(np.arange(N) - N/2)/N
```

```python
fig = plt.figure(figsize=[12,4])

plt.subplot(1,2,1)
plt.plot(omg,np.abs(Xomg))
plt.plot(wn,np.abs(Xfft),'o')
plt.xlabel('$\omega$ [rad]')
plt.title('$|X(e^{j\omega})|$')
plt.grid()

plt.subplot(1,2,2)
plt.plot(omg,np.angle(Xomg))
plt.plot(wn,np.angle(Xfft),'o')
plt.xlabel('$\omega$ [rad]')
plt.title('$\\angle X(e^{j\omega})$')
plt.grid()
```

```python
fig = plt.figure()
ax = fig.subplots()
circ = plt.Circle((0, 0), radius=1, edgecolor='k', facecolor='None')
ax.add_artist(circ)
ax2.set_xlabel('$\mathrm{Re}\{z\}$')
ax2.set_ylabel('$\mathrm{Im}\{z\}$')
ax.grid()
ax.set_aspect(1)

plt.plot(np.cos(wn),np.sin(wn),'d')
```
