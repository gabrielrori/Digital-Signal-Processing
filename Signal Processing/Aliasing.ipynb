# Efecto Aliasing
Aliasing es el efecto por el cual señales continuas en el tiempo, una vez muestreadas pueden llegar a ser indistinguibles. Aliasing puede generar distorsión de las señales y pérdida de la información que contienen.

```python
import numpy as np
import matplotlib.pyplot as plt
#%matplotlib inline
```

```python
# Señales sinusoidales
f1 = 400
f2 = 900
t = np.arange(1E5) * 1E-5
signal1 = np.sin(2*np.pi*f1*t)
signal2 = np.sin(2*np.pi*f2*t)

# Señales muestreadas
fs = 2000
Ns = 2000
n = np.arange(Ns)
tn = n / fs
samples1 = np.sin(2*np.pi*f1*tn)
samples2 = np.sin(2*np.pi*f2*tn)
```

```python
plt.figure(1)
plt.plot(t*1E3,signal1,label='Freq '+str(f1)+' Hz')
plt.plot(t*1E3,signal2,label='Freq '+str(f2)+' Hz')
plt.plot(tn*1E3,samples1,'o')
plt.plot(tn*1E3,samples2,'o')
plt.xlim([0,5.5])
plt.grid('on')
plt.xlabel('Time [ms]')
plt.ylabel('Signals')
plt.legend(loc='lower right')
plt.show()
```

```python
# Análisis en frecuencia
fft_samples1 = np.fft.fftshift(np.fft.fft(samples1))
fft_samples2 = np.fft.fftshift(np.fft.fft(samples2))
spc1 = np.abs(fft_samples1)**2
spc2 = np.abs(fft_samples2)**2
freq = (np.arange(Ns) - Ns/2) * fs / Ns
```

```python
plt.figure(2)
plt.plot(freq,spc1,label='Freq '+str(f1)+' Hz')
plt.plot(freq,spc2,label='Freq '+str(f2)+' Hz')
plt.xlim([-fs/2,fs/2])
plt.grid('on')
plt.xlabel('Frequency [Hz]')
plt.legend(loc='upper left')
plt.show()
```

```python
# Señales muestreadas
fs = 1500
Ns = 1500
n = np.arange(Ns)
tn = n / fs
samples1 = np.sin(2*np.pi*f1*tn)
samples2 = np.sin(2*np.pi*f2*tn)

# Análisis en frecuencia
fft_samples1 = np.fft.fftshift(np.fft.fft(samples1))
fft_samples2 = np.fft.fftshift(np.fft.fft(samples2))
spc1 = np.abs(fft_samples1)**2
spc2 = np.abs(fft_samples2)**2
freq = (np.arange(Ns) - Ns/2) * fs / Ns
```

```python
plt.figure(3)
plt.plot(freq,spc1,label='Freq '+str(f1)+' Hz')
plt.plot(freq,spc2,label='Freq '+str(f2)+' Hz')
plt.xlim([-fs/2,fs/2])
plt.grid('on')
plt.xlabel('Frequency [Hz]')
plt.legend(loc='upper left')
plt.show()
```

```python
f3 = fs - f2
signal3 = np.sin(2*np.pi*f3*t)

plt.figure(4)
plt.plot(t*1E3,signal1,label='Freq '+str(f1)+' Hz')
plt.plot(t*1E3,signal2,label='Freq '+str(f2)+' Hz')
plt.plot(tn*1E3,samples1,'o')
plt.plot(tn*1E3,samples2,'o')
plt.plot(t*1E3,signal3,'--',label='Freq '+str(np.abs(f3))+' Hz')
plt.xlim([0,5.5])
plt.grid('on')
plt.xlabel('Time [ms]')
plt.ylabel('Signals')
plt.legend(loc='lower right')
```
