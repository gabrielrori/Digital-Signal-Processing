# Transformada Discreta de Fourier
Convolución circular vs. convolución lineal

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 16})
#%matplotlib widget
#%matplotlib inline
```

```python
# Señales x1 y x2
N = 8
n = np.arange(N)

x1 = np.array([8,7,6,5,4,3,2,1])
x2 = np.array([1,1,1,1,0,0,0,0])

fig, ax = plt.subplots()
ax.stem(n,x1)
ax.set_title('$x_1[n]$')
ax.grid('on')

fig, ax = plt.subplots()
ax.stem(n,x2)
ax.set_title('$x_2[n]$')
ax.grid('on')
```

```python
# Convolución circular vs. convolución lineal
Xk1 = np.fft.fft(x1)
Xk2 = np.fft.fft(x2)
Yk = Xk1 * Xk2
yc = np.fft.ifft(Yk)

yl = np.convolve(x1,x2)
nl = np.arange(np.size(yl))

fig, ax = plt.subplots()
ax.stem(n,np.real(yc))
ax.set_title('Circular convolution')
ax.grid('on')

fig, ax = plt.subplots()
ax.stem(nl,np.real(yl))
ax.set_title('Linear convolution')
ax.grid('on')
```

```python
x1_b = np.concatenate((x1,np.zeros(N)))
x2_b = np.concatenate((x2,np.zeros(N)))

Xk1_b = np.fft.fft(x1_b)
Xk2_b = np.fft.fft(x2_b)
Yk_b = Xk1_b * Xk2_b
y_b = np.fft.ifft(Yk_b)

n_b = np.arange(np.size(y_b))

fig, ax = plt.subplots()
ax.stem(n_b,np.real(y_b))
ax.set_title('Linear convolution using DFT')
ax.grid('on')

fig, ax = plt.subplots()
ax.stem(nl,np.real(yl))
ax.set_title('Linear convolution')
ax.grid('on')
```
