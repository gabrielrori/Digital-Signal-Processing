# Transformada Discreta de Fourier
Analizando la dispersión en frecuencia y la perdida de información temporal

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 16})
#%matplotlib widget
#%matplotlib inline
```

```python
# Dispersión en frecuencia
N = 16

k1 = 5
k2 = 5.1

fn1 = k1/N
fn2 = k2/N
n = np.arange(N)

x1 = np.cos(2*np.pi*fn1*n)
x2 = np.cos(2*np.pi*fn2*n)

Ns = 1024
tn = np.arange(Ns)* N / Ns
signal1 = np.cos(2*np.pi*fn1*tn)
```

```python
fig, ax = plt.subplots()
ax.stem(n,x1)
ax.plot(tn,signal1,alpha=0.5)
ax.set_xlim([0,N-1])
ax.set_title('$x_1[n] = \cos(2\pi \cdot '+str(k1)+' \cdot n/N)$')
ax.grid('on')

fig, ax = plt.subplots()
ax.stem(n,x2)
ax.set_xlim([0,N-1])
ax.set_title('$x_2[n] = \cos(2\pi \cdot '+str(k2)+' \cdot n/N)$')
ax.grid('on')
```

```python
Xf1 = np.fft.fft(x1)
Xf2 = np.fft.fft(x2)

k = np.arange(N)

fig, ax = plt.subplots()
ax.stem(k,np.abs(Xf1))
ax.set_xlim([0,N-1])
ax.set_title('$X_1[k] = \mathrm{DFT}\{x_1[n]\}$')
ax.grid('on')

fig, ax = plt.subplots()
ax.stem(k,np.abs(Xf2))
ax.set_xlim([0,N-1])
ax.set_title('$X_2[k] = \mathrm{DFT}\{x_2[n]\}$')
ax.grid('on')
```

```python
# Perdida de información temporal

N = 2048
N2 = 1024
n = np.arange(N)
k1 = 6.5
k2 = 24.5
y1 = np.sin(2*np.pi*k1*n/N)
y2 = np.sin(2*np.pi*k2*n/N)
x1 = np.concatenate((y1[0:N2],y2[N2:N]))
x2 = 0.65 * (y1+y2)
```

```python
fig, ax = plt.subplots(figsize=[8,4])
ax.plot(n,x1)
ax.set_xlim([0,N-1])
ax.set_title('$x_1[n]$')
ax.grid('on')

fig, ax = plt.subplots(figsize=[8,4])
ax.plot(n,x2)
ax.set_xlim([0,N-1])
ax.set_title('$x_2[n]$')
ax.grid('on')
```

```python
Xk1 = np.fft.fft(x1)
Xk2 = np.fft.fft(x2)
k = np.arange(N)
```

```python
fig, ax = plt.subplots(figsize=[8,4])
ax.stem(k,np.abs(Xk1))
ax.set_xlim([0,32])
ax.set_title('$|X_1[k]|$')
ax.grid('on')

fig, ax = plt.subplots(figsize=[8,4])
ax.stem(n,np.abs(Xk2))
ax.set_xlim([0,32])
ax.set_title('$|X_2[k]|$')
ax.grid('on')
```

```python
fig, ax = plt.subplots(figsize=[8,4])
ax.stem(k,np.angle(Xk1))
ax.set_xlim([0,32])
ax.set_title('$\\angle X_1[k]$')
ax.grid('on')

fig, ax = plt.subplots(figsize=[8,4])
ax.stem(n,np.angle(Xk2))
ax.set_xlim([0,32])
ax.set_title('$\\angle X_2[k]$')
ax.grid('on')
```
