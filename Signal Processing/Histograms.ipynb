# Histogramas

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
plt.rcParams['font.size'] = '12'
#%matplotlib inline
```

```python
# Histograma de una distribución Gaussiana

N = 2000
x = np.random.randn(N)
n = np.arange(N)
xrange = [-3,3]

Nb = 13
dx = (xrange[1] - xrange[0]) / (Nb-1)
bins = xrange[0]-dx/2 + np.arange(Nb+1)*dx
cbin = xrange[0] + np.arange(Nb)*dx

plt.figure()
plt.plot(n,x,'-o')
plt.grid('on')
plt.xlabel('$n$')
plt.ylabel('$x[n]$')
plt.ylim((bins[0]*1.1,bins[-1]*1.1))

plt.figure()
plt.hist(x,bins=bins,orientation='horizontal',rwidth=0.8,align='mid',ec='k')
plt.grid('on')

plt.figure()
plt.hist(x,bins=bins,rwidth=0.8,align='mid',ec='k')
plt.grid('on')

plt.figure()

h,_,_ = plt.hist(x,bins=bins,rwidth=0.8,align='mid',cumulative=True, ec='k')
plt.plot(cbin,h,'ro-')
plt.grid('on')
```

```python
# Analizando la señal en el dominio de la frecuencia
Xw = np.fft.fftshift(np.fft.fft(x))
w = 2*np.pi * (np.arange(N)-N/2)/N

plt.figure()
plt.plot(w,np.abs(Xw)**2)
plt.grid('on')
plt.xlabel('$\omega$ [rad]')
plt.ylabel('$|X(\omega)|^2$')
```

```python
M = 20
Nfft = np.intc(N/M)
xm = np.reshape(x,[Nfft,M])
Xm = np.fft.fftshift(np.fft.fft(xm,axis=0),axes=0)
wm = 2*np.pi*(np.arange(Nfft) - Nfft/2) / Nfft

X2 = np.mean(np.abs(Xm)**2,axis=1)

plt.figure()
plt.plot(wm,np.abs(Xm)**2)
plt.grid('on')
plt.xlabel('$\omega$ [rad]')
plt.ylabel('$|X(\omega)|^2$')

plt.figure()
plt.plot(wm,X2)
plt.grid('on')
plt.xlabel('$\omega$ [rad]')
plt.ylabel('$|X(\omega)|^2$')
plt.ylim([-0.1,np.max(X2)*1.1])

```

```python
# Histograma de una imagen

img = mpimg.imread('JRO-view.jpg')

plt.figure()
plt.imshow(img)

sz = np.shape(img)

img = img.reshape([sz[0]*sz[1],sz[2]])

r = img[:,0]
g = img[:,1]
b = img[:,2]

plt.figure()
plt.hist(r,bins=256,range=[-0.5,255.5],color='r')
plt.grid('on')

plt.figure()
plt.hist(g,bins=256,range=[-0.5,255.5],color='g')
plt.grid('on')

plt.figure()
plt.hist(b,bins=256,range=[-0.5,255.5],color='b')
plt.grid('on')
```
