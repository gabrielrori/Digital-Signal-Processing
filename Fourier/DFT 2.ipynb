# Transformada Discreta de Fourier
Periodicidad y resolución en frecuencia

```python
import numpy as np
import matplotlib.pyplot as plt
#%matplotlib widget
#%matplotlib inline
```

```python
# Señal Chirp

B = 400

fs = 1024
Ns = 1024

n = np.arange(Ns)
t = n / fs

phase = np.pi * B * t**2

x = (Ns-n)/Ns * np.cos(phase)
#x = np.cos(phase)

```

```python
fig, ax = plt.subplots()
ax.plot(n,x)
ax.grid('on')
ax.set_xlim([0,Ns-1])
```

```python
delta_f = fs / Ns
k = np.arange(Ns)
freq = k * delta_f

Xk = np.fft.fft(x)

plt.figure()
plt.plot(freq,np.abs(Xk))
plt.xlim([0,fs])
plt.xlabel('Frequency [Hz]')
plt.grid(True)
```

```python
freq = (k - Ns/2) * delta_f

Xk_shift = np.fft.fftshift(Xk)

plt.figure()
plt.plot(freq,np.abs(Xk_shift))
plt.xlim([-fs/2,fs/2])
plt.xlabel('Frequency [Hz]')
plt.grid(True)
```
