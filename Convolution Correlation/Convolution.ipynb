# Convolution
Un sistema discreto lineal e invariante en el tiempo (LTI) está completamente caracterizado por su respuesta impulsiva $h[n]$.

$$
\overset{x[n]}{\longrightarrow}
{\boxed{\: h[n] \:}}
\overset{x_L[n]}{\longrightarrow}
$$

Dado $h[n]$ es posible calcular $y[n]$, para toda entrada $x[n]$, aplicando la función de convolución discreta,

$$
y[n] = x[n] * h[n] = \sum_{k=-\infty}^{\infty} x[k] h[n-k] .
$$


```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 24})
#%matplotlib inline
```

```python
h = np.array([1,2,1,-1])
x = np.array([1,2,3,1])

k_h = np.arange(-1,3)
k_x = np.arange(0,4)
```

```python
fig, ax1 = plt.subplots()
ax1.stem(k_h,h)
ax1.set_xticks(k_h)
ax1.set_xlabel('$k$')
ax1.set_ylabel('$h[k]$')
ax1.grid()

fig, ax2 = plt.subplots()
ax2.stem(k_x,x)
ax2.set_xticks(k_x)
ax2.set_xlabel('$k$')
ax2.set_ylabel('$x[k]$')
ax2.grid()
```

```python
n = 2

hflip = np.flip(h)
k_h2 = np.flip(-k_h)

fig, ax3 = plt.subplots()
ax3.stem(k_h2,hflip)
ax3.set_xticks(k_h2)
ax3.set_xlabel('$k$')
ax3.set_ylabel('$h[-k]$')
ax3.grid()

k_h3 = n + k_h2
fig, ax4 = plt.subplots()
ax4.stem(k_h3,hflip)
ax4.set_xticks(k_h3)
ax4.set_xlabel('$k$')
ax4.set_ylabel('$h['+str(n)+'-k]$')
ax4.grid()

```

```python
xh = hflip * x

fig, ax5 = plt.subplots()
ax5.stem(k_x,xh)
ax5.set_xticks(k_x)
ax5.set_xlabel('$k$')
ax5.set_ylabel('$x[k] \, h['+str(n)+'-k]$')
ax5.grid()
```

```python
y = np.convolve(x,h)
n_y = np.arange(-1,6)

print(y)

fig, ax6 = plt.subplots()
ax6.stem(n_y,y)
ax6.set_xticks(n_y)
ax6.set_yticks(np.arange(np.min(y),np.max(y)+1,2))
ax6.set_xlabel('$n$')
ax6.set_ylabel('$y[n]$')
ax6.grid()
```
