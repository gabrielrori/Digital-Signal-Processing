# Transformada Z - Unit step

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
#%matplotlib widget
%matplotlib inline
```

```python
# Ejemplo X(z) = z / z-a
a = 1

x = np.linspace(-3,3,100)
y = np.linspace(-3,3,100)

xx, yy = np.meshgrid(x,y)
zz = xx+1j*yy

ROC = np.abs(zz) < a

Xz = np.polyval([1],zz)/np.polyval([1,-a],zz)
Xz[ROC] = np.NaN
```

```python
fig = plt.figure()
ax = fig.subplots(subplot_kw={"projection": "3d"})
surf = ax.plot_surface(xx,yy,np.abs(Xz))
#surf = ax.plot_surface(xx,yy,np.abs(Xz),cmap=cm.coolwarm,antialiased=False)
ax.set_xlabel('$\mathrm{Re}\{z\}$')
ax.set_ylabel('$\mathrm{Im}\{z\}$')
ax.set_zlabel('$|X(z)|$')
```

```python
fig = plt.figure()
ax = fig.subplots(subplot_kw={"projection": "3d"})
surf = ax.plot_surface(xx,yy,np.angle(Xz))
#surf = ax.plot_surface(xx,yy,np.unwrap(np.angle(Xz),axis=0))
ax.set_xlabel('$\mathrm{Re}\{z\}$')
ax.set_ylabel('$\mathrm{Im}\{z\}$')
ax.set_zlabel('$\\angle X(z)$')
```
