# Transformada Discreta de Fourier
Interpretaci칩n de la DFT como medida de la correlaci칩n con se침ales cosenoidales y senoidales para diferentes frecuencias.

```python
import numpy as np
import matplotlib.pyplot as plt
#%matplotlib widget
#%matplotlib inline
```

```python
# DFT de una se침al cosenoidal

N = 16
n = np.arange(N)
omg = 2*np.pi*0.125
#omg = 2*np.pi*0.1
x = np.cos(omg*n)

Nt = 1024
t = np.arange(Nt) * N /Nt
xt = np.cos(omg*t)

k = 2
cosk = np.cos(2*np.pi*n*k/N)
sink = np.sin(2*np.pi*n*k/N)

x_cosk = x * cosk
x_sink = x * sink

print('X[{0:.0f}] = {1:.2f} - {2:.2f}j'.format(k,np.sum(x_cosk),np.sum(x_sink)))
```

```python
fig, axs = plt.subplots(3,2,sharex=True,figsize=[15,9])

ax = axs[0,0]
ax.stem(n,x)
ax.plot(t,xt,'k',alpha=0.3)
ax.set_title('Secuencia de entrada')
ax.grid('on')
ax.set_xlim([0,N-1])

ax = axs[0,1]
ax.stem(n,x)
ax.plot(t,xt,'k',alpha=0.3)
ax.set_title('Secuencia de entrada')
ax.grid('on')

ax = axs[1,0]
ax.stem(n,cosk)
ax.set_title('$\cos(2 \pi \cdot k \cdot n / N), \; k='+str(k)+', \; n \in [0,'+str(N-1)+'] $')
ax.grid('on')

ax = axs[1,1]
ax.stem(n,sink)
ax.set_title('$\sin(2 \pi \cdot k \cdot n / N), \; k='+str(k)+', \; n \in [0,'+str(N-1)+'] $')
ax.grid('on')

ax = axs[2,0]
ax.stem(n,x_cosk)
ax.set_title('$x \cdot \cos(2 \pi \cdot '+str(k)+' \cdot n / N) $')
ax.set_xlabel('$n$')
ax.grid('on')

ax = axs[2,1]
ax.stem(n,x_sink)
ax.set_title('$x \cdot \sin(2 \pi \cdot '+str(k)+' \cdot n / N) $')
ax.set_xlabel('$n$')
ax.grid('on')
```

```python
# FFT de una se침al cosenoidal
Xk = np.fft.fft(x)

fig, ax = plt.subplots()

ax.stem(n,np.abs(Xk))
ax.set_title('$X[k] = \mathrm{DFT}\{x[n]\}$')
ax.set_xlim([0,N-1])
ax.set_xlabel('$k$')
ax.grid('on')
```
